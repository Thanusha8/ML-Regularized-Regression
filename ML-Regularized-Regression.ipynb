```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet 
from sklearn.metrics import mean_squared_error 
```

```python
data = pd.read_csv ("C:\\Thanusha\\Python\\ML master class\\Regularization\\Data\\Boston.csv")
```

```python
data.head()
```

```python
x= data.iloc[:,:12].values
y= data.iloc[:,12].values
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

```python
m1= LinearRegression()
m2= Ridge(alpha=1)
m3= Lasso(alpha=1)
m4= ElasticNet(alpha=1)
```

```python
m1.fit(x_train,y_train)
m2.fit(x_train,y_train)
m3.fit(x_train,y_train)
m4.fit(x_train,y_train)
```

```python
m1.coef_
```

```python
m2.coef_
```

```python
m3.coef_
```

```python
m4.coef_
```

```python
yp1 = m1.predict(x_test)
yp2 = m2.predict(x_test)
yp3 = m3.predict(x_test)
yp4 = m4.predict(x_test)
```

```python
print(np.sqrt(mean_squared_error(y_test,yp1)))
print(np.sqrt(mean_squared_error(y_test,yp2)))
print(np.sqrt(mean_squared_error(y_test,yp3)))
print(np.sqrt(mean_squared_error(y_test,yp4)))
```
